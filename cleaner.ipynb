```python
import pandas as PD
import numpy as NP
```

```python
df_orig:PD.DataFrame = PD.read_csv("data/application_data.csv")
```

In the code block below, these are selected because of their relative abundance of non-null values.

```python
# Finding out the percentage of missing values in each column in the given dataset:
percent_missing_app = round(100*(df_orig.isnull().mean()), 2)
#Get the columns having greater than 50% null value for application data
percent_missing_app = percent_missing_app[percent_missing_app.values > 25.0]
#Removing those columns which have missing values more than 25% in application data
cols_dropped=([percent_missing_app[percent_missing_app >=25.00].index])
df_dropped = df_orig.drop(labels=cols_dropped[0],axis=1)
```

```python
df_dropped.columns
```

```python
df_dropped = df_dropped.dropna()
df_dropped = df_dropped.drop("SK_ID_CURR", axis=1)
```

```python
numerical_df = df_dropped.select_dtypes(include='number')
categorical_col = []
for col in numerical_df:
    if col == "TARGET": continue
    unique_values = len(numerical_df[col].unique())
    print(col, "---", unique_values)
    if unique_values < 4:
        categorical_col.append(col)
```

```python
df_continuous = numerical_df.drop(categorical_col, axis = 1)
for col in df_continuous:
    if col == "TARGET": continue
    unique_values = len(df_continuous[col].unique())
    print(col, "---", unique_values)
```

```python
df_continuous.to_csv("cleaned_continuous.csv", index=False)
```

```python
# Normalisation
df_normalised:PD.DataFrame = PD.DataFrame()
for colName in df_continuous:
    if colName == "TARGET":
        df_normalised[colName] = df_continuous[colName]
        continue
    col:PD.Series = df_continuous[colName]
    minimum = col.min()
    maximum = col.max()
    valueRange = maximum - minimum
    df_normalised[colName] = (col - minimum) / valueRange
df_normalised
```

```python
df_normalised.reset_index(inplace=True)
df_normalised.to_csv("cleaned_continuous_normalised.csv", index=False)
```

```python
df_categorical:PD.DataFrame = df_dropped.copy()
for col in df_continuous:
    if col == "TARGET": continue
    df_categorical = df_categorical.drop(col, axis=1)
df_categorical
```

```python
for col in df_categorical:
    if col == "TARGET": continue
    unique_values = len(df_categorical[col].unique())
    print(col, "---", unique_values)
```

```python
df_categorical = df_categorical.drop("FLAG_MOBIL",axis=1)
df_categorical.reset_index(inplace=True, drop=True)
```

```python
df_categorical.drop("FLAG_DOCUMENT_2", axis=1, inplace=True)
```

```python
df_categorical.to_csv("cleaned_categorical.csv", index=False)
```

```python
# Manual one hot encoding
df_categorical_final = PD.DataFrame()
for colName in df_categorical:
    if colName == "TARGET": 
        df_categorical_final[colName] = df_categorical[colName]
        continue
    col = df_categorical[colName]
    if col.unique().__len__() == 2 and col.dtype == NP.int64:
        df_categorical_final[colName] = col
        continue
    onehot = PD.get_dummies(df_categorical[colName]).rename(lambda x: colName + "_" + str(x), axis= 1)
    df_categorical_final = PD.concat([df_categorical_final, onehot], axis = 1)
df_categorical_final
```

```python
df_categorical_final.to_csv("cleaned_categorical_onehot.csv", index=False)
```
