## Logistic Regression

We try to predict defaulters here using `logistic regression`. A logistic regression model is quite suitable here since we only have 2 classes to classify the data into. That is, defaulters and non defaulters.

```python
import os
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
def split_data(data):
    x = data.iloc[:, 1:]
    y = data['TARGET']
    x_train, x_rem, y_train, y_rem = train_test_split(x, y, random_state=0, test_size=0.2)
    x_validation, x_test, y_validation, y_test = train_test_split(x_rem, y_rem, random_state=0, test_size=0.5)
    return x_train, x_test, x_validation, y_train, y_test, y_validation
```

```python
def load_model(file_name):
    loaded_model = pickle.load(open(file_name, 'rb'))
    return loaded_model
```

```python
def save_model(model, file_name):
    pickle.dump(model, open(file_name, 'wb'))
```

```python
def custom_score(model, x_valid, y_valid):
    tp_weight = 0.6
    y_pred = model.predict(x_valid)
    pred_non_defaulters = y_pred[y_valid == 0]
    pred_defaulters = y_pred[y_valid == 1]
    true_pos_rate = pred_defaulters.sum() / pred_defaulters.size
    true_neg_rate = 1 - pred_non_defaulters.sum() / pred_non_defaulters.size
    return true_neg_rate * (1 - tp_weight) + true_pos_rate * tp_weight
```

```python
# Train logistic regression model with hyperparam tuning
def train_log_reg_model(x_train, x_validation, y_train, y_validation):
    solver_penalties = {
        'liblinear': ['l1', 'l2'],
        'sag': ['l2', None],
        'saga': ['l1', 'l2', 'elasticnet', None],
        'lbfgs': ['l2', None],
    }
    optimal_model = None
    max_score = float('-inf')
    for solver, penalties in solver_penalties.items():
        for penalty in penalties:
            n_jobs = None if solver == 'liblinear' else -1
            l1_ratios = [i/10 for i in range(1,10)] if penalty == 'elasticnet' else [None]
            regularization_strengths = [10 ** (i - 3) for i in range(7)] if penalty is not None else [1.0]
            for c in regularization_strengths:
                for l1_ratio in l1_ratios:
                    logistic_regression_model = LogisticRegression(solver=solver, penalty=penalty, C=c,
                                                                   random_state=0, max_iter=3000, 
                                                                   n_jobs=n_jobs, l1_ratio=l1_ratio)
                    logistic_regression_model.fit(x_train, y_train)
                    score = logistic_regression_model.score(x_validation, y_validation)
                    if score > max_score:
                        max_score = score
                        optimal_model = logistic_regression_model
    return optimal_model
                    
```

```python
def plot_cm(model, x_test, y_test, data_used=None):
    class_labels = ['Non Defaulter', 'Defaulter']
    disp = ConfusionMatrixDisplay.from_estimator(model, x_test, y_test,
                                                 cmap="Blues", display_labels=class_labels)
    score = round(model.score(x_test, y_test) * 100, 2)
    true_positives = float(disp.text_[1][1].get_text())
    true_negatives = float(disp.text_[0][0].get_text())
    false_positives = float(disp.text_[0][1].get_text())
    false_negatives = float(disp.text_[1][0].get_text())
    total_non_defaulters = true_negatives + false_positives
    total_defaulters = true_positives + false_negatives
    true_positive_rate = round((true_positives / total_defaulters) * 100, 2)
    false_negative_rate = round((false_negatives / total_defaulters) * 100, 2)
    true_negative_rate = round((true_negatives / total_non_defaulters) * 100, 2)
    false_positive_rate = round((false_positives / total_non_defaulters) * 100, 2)
    model_type = 'full' if data_used is None else data_used
    disp.ax_.set_title(f'Confusion matrix of logistic regression model ({model_type})\n'
                       + f'Accuracy score: {score}%\n' 
                       + f'True positive rate: {true_positive_rate}%\n' 
                       + f'False negative rate: {false_negative_rate}%\n' 
                       + f'True negative rate: {true_negative_rate}%\n' 
                       + f'False positive rate: {false_positive_rate}%')
    plt.show()
```

```python
# Read cleaned data
data = pd.read_csv("data/cleaned_final.csv")
```

```python
# Sample 100% of data (full data)
full_shuffled = data.sample(frac=1)
full_shuffled.reset_index(drop=True, inplace=True)
full_shuffled
```

```python
# Split the data into 80% training set, 10% test set and 10% validation set
(x_train_full, x_test_full, x_validation_full,
 y_train_full, y_test_full, y_validation_full) = split_data(full_shuffled)

print(x_train_full.shape, x_test_full.shape, x_validation_full.shape, 
      y_train_full.shape, y_test_full.shape, y_validation_full.shape)
```

```python
# load full trained model if exists, else train new logistic regression model
full_model_file_name = 'log_reg_predictor_full.sav'
path = os.path.join(os.getcwd(), full_model_file_name)
trained_model_full = None
if os.path.exists(path):
    trained_model_full = load_model(full_model_file_name)
else:
    trained_model_full = train_log_reg_model(x_train_full, x_validation_full,
                                             y_train_full, y_validation_full)
    save_model(trained_model_full, full_model_file_name)
```

```python
# Show full trained model and its accuracy
print("Trained Model Full:", trained_model_full)
print("Accuracy:", trained_model_full.score(x_test_full, y_test_full))
```

```python
plot_cm(trained_model_full, x_test_full, y_test_full)
```

As seen above, the true positive rate is very low. That is, the accuracy of the model in predicting a defaulter is very low. This is likely due to an under-representation of defaulters in the dataset (there is not enough data on defaulters to accurately predict a defaulter). Hence, let's try using a dataset that has an equal ratio of defaulters and non defaulters and see if this can potentially improve the true positive rate.

```python
# create a subset of data so that defaulters and non-defaulters are 1:1
defaulters = data.loc[data['TARGET'] == 1]
non_defaulters = data.loc[data['TARGET'] == 0]
downsampled_non_defaulters = non_defaulters.sample(len(defaulters))
undersampled = pd.concat([downsampled_non_defaulters, defaulters])
undersampled.reset_index(drop=True, inplace=True)
undersampled
```

```python
# Split the data into 80% training set, 10% test set and 10% validation set
(x_train_undersampled, x_test_undersampled, x_validation_undersampled,
 y_train_undersampled, y_test_undersampled, y_validation_undersampled) = split_data(undersampled)

print(x_train_undersampled.shape, x_test_undersampled.shape, x_validation_undersampled.shape, 
      y_train_undersampled.shape, y_test_undersampled.shape, y_validation_undersampled.shape)
```

```python
# load undersampled trained model if exists, else train new logistic regression model
undersampled_model_file_name = 'log_reg_predictor_undersampled.sav'
path = os.path.join(os.getcwd(), undersampled_model_file_name)
trained_model_undersampled = None
if os.path.exists(path):
    trained_model_undersampled = load_model(undersampled_model_file_name)
else:
    trained_model_undersampled = train_log_reg_model(x_train_undersampled, x_validation_undersampled,
                                               y_train_undersampled, y_validation_undersampled)
    save_model(trained_model_undersampled, undersampled_model_file_name)
```

```python
# Show model and model accuracy
print("Trained Model Undersampled:", trained_model_undersampled)
print("Accuracy:", trained_model_undersampled.score(x_test_undersampled, y_test_undersampled))
```

```python
plot_cm(trained_model_undersampled, x_test_undersampled, y_test_undersampled, data_used='undersampled')
```

As seen from the results above, the true positive rate drastically improved from `0.43%` to `67.32%`. However, we also see a decrease in true negative rate from `99.97%` to `70.12%` and a drop in accuracy from `92.35%` to `68.69%`. Accuracy and true negative rate are now lower because the dataset used to train this model contains a lot less data on non defaulters compared to the one used to train the previous model. Nevertheless, this model can still predict, with reasonable accuracy (around 70% true negative and true positive rate), if someone is likely to be a defaulter or not. Hence, this model may be more useful than the previous one since the previous model is only very good at predicting non defaulters but not defaulters.

Now, having seen the improved results from the under sampling (removing non defaulter data until number of defaulters and non defaulters are the same, lets now try oversampling to see how the results will differ. That is, we randomly duplicate the number of defaulters until the number of defaulters and non defaulters are equal.

```python
diff = len(non_defaulters) / len(defaulters)
upsampled_defaulters = defaulters.sample(frac=diff, replace=True)
oversampled = pd.concat([non_defaulters, upsampled_defaulters])
oversampled.reset_index(drop=True, inplace=True)
oversampled
```

```python
# Split the data into 80% training set, 10% test set and 10% validation set
(x_train_oversampled, x_test_oversampled, x_validation_oversampled,
 y_train_oversampled, y_test_oversampled, y_validation_oversampled) = split_data(oversampled)

print(x_train_oversampled.shape, x_test_oversampled.shape, x_validation_oversampled.shape,
 y_train_oversampled.shape, y_test_oversampled.shape, y_validation_oversampled.shape)
```

```python
# load oversampled trained model if exists, else train new logistic regression model
oversampled_model_file_name = 'log_reg_predictor_oversampled.sav'
path = os.path.join(os.getcwd(), oversampled_model_file_name)
trained_model_oversampled = None
if os.path.exists(path):
    trained_model_oversampled = load_model(oversampled_model_file_name)
else:
    trained_model_oversampled = train_log_reg_model(x_train_oversampled, x_validation_oversampled,
                                               y_train_oversampled, y_validation_oversampled)
    save_model(trained_model_oversampled, oversampled_model_file_name)
```

```python
# Show model and model accuracy
print("Trained Model Oversampled:", trained_model_oversampled)
print("Accuracy:", trained_model_oversampled.score(x_test_oversampled, y_test_oversampled))
```

```python
plot_cm(trained_model_oversampled, x_test_oversampled, y_test_oversampled, data_used='oversampled')
```

As seen from the confusion matrix above, its accuracy of the oversampled model (`68.61%`) is largely similar to that of the undersampled model (`68.69%`). It is also slightly better at predicting defaulters than the undersampled model as it has a higher true positive rate of `68.16%` compared to the undersampled model's true positivate rate of `67.32%`. Finally, this model is slightly worse at predicting non defaulters than the undersampled model as it has a lower true negative rate of `69.06%` compared to the undersampled model's true negative rate of `70.12%`.

We believe that false negatives (where the model predicts a person is a non defaulter when he actually is a defaulter) is more costly than false positives (where the model predicts a person is a defaulter when he actually is not) to a company as there is a high risk of default if a company were to loan a false negative customer money. Resulting in huge losses for the company. On the other hand, not loaning money to a false positive customer, will only result in the company losing a potential customer. Thus, the costs of false negatives will definitely be much higher than that of false positives.

Therefore, our goal will be to reduce the number of false negatives as much as possible, or equivalently, maximise the true positive rate. Hence, based on the results we are getting, we will choose the oversampled model which gives the highest true postive rate as our model of choice here.
